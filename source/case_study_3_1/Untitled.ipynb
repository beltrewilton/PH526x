```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import numpy as np
import pandas as pd
```

```python
fig, ax = plt.subplots(1, 1)
print(1)
```

```python
X = np.linspace(norm.ppf(.01), norm.ppf(.99), 1000)
ax.plot(X, norm.pdf(X), 'r-', lw=5, alpha=.6, label='norm pdf')
```
